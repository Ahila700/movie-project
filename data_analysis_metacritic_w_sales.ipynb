```python
import requests 
import pandas as pd 
import json 
import numpy as np

from bs4 import BeautifulSoup
```

```python
# importing csv files for game sales and metacritic reviews
```

```python
df = pd.read_csv('metacritic_reviews.csv', index_col = 0)
```

```python
df.shape
```

```python
dfs = pd.read_csv('Game_sales.csv', index_col = 0)
```

```python
dfs.shape
```

```python
#cleaning up some of the columns prior to merging 
```

```python
dfs = dfs.drop(columns = 'Genre')
```

```python
df = df.rename(columns = {'title': 'Game', 'developer': 'Publisher'})
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x))

genre_list = []
for x in df.genres:
    genre_list.append(x[0].split(','))
    
df['genres'] = genre_list
```

```python
# putting together the 2 dataframes where the name of the game is the same so it is 
# possible to analyze data 
```

```python
metacritic_w_sales = pd.merge(df, dfs, on = 'Game')
```

```python
metacritic_w_sales
```

```python
metacritic_w_sales = metacritic_w_sales.drop(columns = 'Publisher_x')
```

```python
metacritic_w_sales = metacritic_w_sales.rename(columns = {'Publisher_y': 'Publisher'})
```

```python
metacritic_w_sales.to_csv('metacritic_v_sales.csv')
```
